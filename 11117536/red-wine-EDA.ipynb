### Exploratory Data Analysis

```python
new_df['quality'].value_counts()
```

```python
# Histogram of a alcohol column
plt.figure(figsize=(8,6))
plt.hist(new_df['alcohol'])
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.title('Alcohol\'s histogram')
```

### Inferences on alcohol histogram

- The red wines in the dataset have alcohol content ranging from 9% to 14%
- There is a peak around 9.5% to 10.5% alcohol content, indicating a relatively high frequency of wines with lower alcohol content
- There is a tail towards higher alcohol content, suggesting the presence of few number wines with higher alcohol levels in the dataset

```python
# Boxplot of a alcohol column
plt.figure(figsize=(8,6)) # set the figure size
plt.boxplot(x=new_df['alcohol'])
plt.title('Boxplot of Alcohol')
plt.ylabel('Alcohol')
plt.show()
```

### Inferences on boxplot of alcohol

- The median alcohol content is around 10.3%
- The box is slightly skewed to the lower end, suggesting that there is a concentration of wines with lower alcohol content
- The whiskers in the plot indicate the range of the data, excluding outliers.

```python
# bar graph of quality column
plt.figure(figsize=(8,6))
quality_counts = new_df['quality'].value_counts()
plt.bar(x=quality_counts.index, height=quality_counts.values)
plt.title('Bar graph of Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
```

### Inferences on bar graph of quality

- The majority of wines in the dataset are rated with a quality rating of 5 or 6
- Wines with quality ratings of 7 and 4 are less common compared to ratings of 5 and 6
- Wines with quality ratings of 3 and 8 are the least common in the dataset.

```python
# scatterplot of ph vs chlorides
plt.figure(figsize=(8,6))
plt.scatter(x=new_df['pH'], y=new_df['chlorides'], s=10, c='blue', alpha=0.7)
plt.title('Scatterplot of pH vs chlorides')
plt.xlabel('pH')
plt.ylabel('chlorides')
plt.show()
```

### Inferences scatterplot of ph vs chlorides

- The majority of the data points are concentrated in a specific range of chlorides and pH values, indicating the presence of a cluster or grouping
- There appears to be a weak negative linear relationship between pH and chlorides. As the chlorides increase, the pH tends to decrease slightly
- Some data points have relatively higher chlorides and lower pH values, suggesting the presence of wines with higher chloride levels and more acidic pH

```python
# scatterplot of alcohol vs volatile acidity
plt.figure(figsize=(8,6)) # set the figure size
plt.scatter(x=new_df['alcohol'], y=new_df['volatile acidity'], s=10, c='green', alpha=0.7)
plt.title('Scatterplot of Alcohol vs Volatile Acidity')
plt.xlabel('Alcohol')
plt.ylabel('Volatile Acidity')
plt.show()
```

### Inferences on scatterplot of alcohol vs volatile acidity

- The data points are scattered across a wide range of volatile acidity and alcohol values, indicating a diverse distribution
- Wines with volatile acidity exceeding 1.0 g/L and alcohol content below 11% were found in the dataset, suggesting presence of low-alcohol wines with higher perceived acidity

```python
corr_matrix = new_df.corr()
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(np.arange(len(corr_matrix.columns)), corr_matrix.columns)
        
plt.title('Correlation Matrix')
plt.show()
```

### Inferences on correlation matrix

- Strong postive correlation between citric acid content and fixed acidity indicating that wines with higer citric acid content have higher fixed acidity
- Alcohol content has a moderately positive correlation with wine quality. Wines with higher alcohol content tend to have higher quality ratings
- Volatile acidity has a moderately negative correlation with wine quality. Wines with lower volatile acidity tend to have higher quality ratings
- There is a moderate negative correlation between volatile acidity and pH. Wines with higher volatile acidity tend to have lower pH values, indicating higher acidity.

```python
# scatterplot of density vs fixed acidity

plt.figure(figsize=(12, 7))
plt.scatter(new_df['density'], new_df['fixed acidity'])
plt.xlabel('Density')
plt.ylabel('Fixed Acidity')
plt.title('Scatter Plot of Density vs Fixed Acidity')
plt.show()
```
